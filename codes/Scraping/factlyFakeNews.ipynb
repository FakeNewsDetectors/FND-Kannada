##Importing libraries

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

##Scraping factly.in


```python
url = "https://factly.in/category/fake-news-kannada/"
req = Request(url , headers = {'User-Agent' : 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage,"html.parser")

soup.prettify()
```

##Scraping fakenews headlines in webpage 1


```python
table = []
table.append(soup.find_all('h2', attrs = {'class':'post-title'}))
```

##Scraping  fakenews headlines in webpages 2-43

```python
for i in range(2,44):
    url = "https://factly.in/category/fake-news-kannada/page/" + str(i)
    req = Request(url , headers = {'User-Agent' : 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    soup = BeautifulSoup(webpage,"html.parser")
    table.append(soup.find_all('h2', attrs = {'class':'post-title'}))
```

##Appending the HTML anchor tags which contains the headlines

```python
lists1 = []
for i in range(0,42):
  for j in range(1,11):
    lists1.append(table[i][j].find('a'))
```

##The extraction of the links within the anchor tag

```python
lists2 = []
for url in lists1:
  url = str(url)
  url = url.split('"')[1]
  lists2.append(url)
```

##The extraction of the blockquotes which contain the headlines

```python
table2 = []
for links in lists2:
  url = links
  req = Request(url , headers = {'User-Agent' : 'Mozilla/5.0'})
  webpage = urlopen(req).read()
  soup = BeautifulSoup(webpage,"html.parser")
  table2.append(soup.find_all('blockquote', attrs = {'class':'wp-block-quote'}))
```

##Removal of erroneous headlines

```python
lists3 = []
temp_list = []
for i in table2:
  for j in i:
    lists3.append(j.getText())
lists3[190] = lists3[0]
lists3[191] = lists3[0]
lists4 =[]
for index,data in enumerate(lists3):
  lists4.append(data.split(":")[1])
for index,i in enumerate(lists4):
  i = i.split(".")[:-1]
  lists4[index] = ".".join(i)
  temp = [lists4[index]]
  temp_list.append(temp)
```

##Final output: CSV file containing fake news headlines generated

```python
with open('factlyFakeNews.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(temp_list)
```
