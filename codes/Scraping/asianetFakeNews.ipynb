##Importing libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import re
```

##Scraping asianetnews.com 

```python
URL = "https://kannada.asianetnews.com/fact-check"
r = requests.get(URL) 
soup = BeautifulSoup(r.content, 'html5lib') 
soup.prettify()
```

##Extracting links from webpage 1


```python
lists = []
final_list = []
table = soup.find_all('div', attrs = {'class':'mobrightcol'})
for element in table:
  lists.append(element.find('a'))
#formatting
for link in lists:
  url = str(link)
  url = url.split('"')[1]
  temp_list=[]
  temp_list.append(url)
  final_list.append(temp_list)
print(final_list)
```

##Extract links from webpages 2-11

```python
list_for_remaining_pages = []
#scraping links
for i in range(2,11):
  URL = "https://kannada.asianetnews.com/fact-check/page-"+ str(i)
  r = requests.get(URL) 
  soup = BeautifulSoup(r.content, 'html5lib')
  table = soup.find_all('div', attrs = {'class':'mobrightcol'})
  for element in table:
    list_for_remaining_pages.append(element.find('a'))
#formatting
for link in list_for_remaining_pages:
  url = str(link)
  url = url.split('"')[1]
  temp_list=[]
  temp_list.append(url)
  final_list.append(temp_list)
print(final_list)
```

##Checkpoint: CSV file containing links generated

```python
with open('asianetnewsLinks.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(final_list)
```

##Reading csv file with links of all news articles

```python
df = pd.read_csv('asianetnewsLinks.csv')
df.columns = ['url']
df['url'] = 'https://kannada.asianetnews.com' + df['url']
```

##Extracting news headlines

```python
fake_news = []
for i in range(len(df)):
  URL = df.loc[i, 'url']
  r = requests.get(URL) 
  soup = BeautifulSoup(r.content, 'html5lib')
  table = soup.find('h1')
  fake_news.append(table)
```

##Extracting strings between h1 tags

```python
final_list=[]
tag = "h1"
reg_str = "<" + tag + ">(.*?)</" + tag + ">"
for i in range(len(fake_news)):
    fake_news[i] = str(fake_news[i])
    res = re.findall(reg_str, fake_news[i])
    temp_list = []
    temp_headline = str(res).split(":")
    if len(temp_headline)==1:
      temp_list.append(temp_headline[0])
    else:
      temp_list.append(temp_headline[1])
    final_list.append(temp_list)
print(final_list)
print("Length: ", len(final_list))
```

##Final Output: CSV file containing news headlines generated

```python
with open('asianetFakeNews.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(final_list)
```
