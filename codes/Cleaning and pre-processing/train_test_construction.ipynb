```python
# reading fake news dataset scraped from prajavani news
prajavani_fake = pd.read_csv('/content/drive/MyDrive/FND_Kannada/data/v2/Annotated/Fake/prajavaniFakeNews.csv')
prajavani_fake.drop("Unnamed: 0", axis = 1, inplace = True)
prajavani_fake.head()
```

```python
# checking size of true news and fake news datasets
print('Asianet True: ', len(asianet_true))
print('News18 True: ', len(news18_true))
print('Prajavani True: ', len(prajavani_true))
print('Publictv True: ', len(asianet_true))
print('Asianet Fake: ', len(asianet_fake))
print('Factly Fake: ', len(factly_fake))
print('Oneindia Fake: ', len(oneindia_fake))
print('Prajavani Fake: ', len(prajavani_fake))
```

```python
# creating empty train and test datasets
df_train = pd.DataFrame()
df_test = pd.DataFrame()
```

```python
# adding 180 true news and 57 fake news instances scraped from asianet to the testing set
df_test_asianet_true = asianet_true.sample(n = 180)
df_test = df_test.append(df_test_asianet_true)
asianet_true = pd.concat([asianet_true, df_test_asianet_true]).drop_duplicates(keep = False)
print(len(asianet_true))

df_test_asianet_fake = asianet_fake.sample(n = 57)
df_test = df_test.append(df_test_asianet_fake)
asianet_fake = pd.concat([asianet_fake, df_test_asianet_fake]).drop_duplicates(keep = False)
print(len(asianet_fake))
```

```python
# adding the true news and fake news scraped from asianet to the training set after removing the common rows from the testing set
df_train = df_train.append(asianet_true)
df_train = df_train.append(asianet_fake)
```

```python
# printing the size of the training and the testing set
print(len(df_test))
print(len(df_train))
```

```python
# adding 100 true news and 15 fake news instances scraped from prajavani to the testing set
df_test_prajavani_true = prajavani_true.sample(n = 100)
df_test = df_test.append(df_test_prajavani_true)
prajavani_true = pd.concat([prajavani_true, df_test_prajavani_true]).drop_duplicates(keep = False) # removing the rows from the testing set
print(len(prajavani_true))

df_test_prajavani_fake = prajavani_fake.sample(n = 15)
df_test = df_test.append(df_test_prajavani_fake)
prajavani_fake = pd.concat([prajavani_fake, df_test_prajavani_fake]).drop_duplicates(keep = False)
print(len(prajavani_fake))
```

```python
# adding the true news and fake news scraped from prajavani to the training set after removing the common rows from the testing set
df_train = df_train.append(prajavani_true)
df_train = df_train.append(prajavani_fake)
```

```python
# printing the size of the training and the testing set
print(len(df_test))
print(len(df_train))
```
