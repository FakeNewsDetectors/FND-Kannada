```python
# importing necessary libraries
import pandas as pd
```

```python
# reading the dataset
df_train = pd.read_csv('/content/drive/MyDrive/FND_Kannada/data/v2/train_raw.csv')
df_train.head()
```

```python
text = df_train['news']
text.head()
```

```python
# function for removing punctuations and irrelavant text
def punctuation_cleaning():
  for i in range(len(text)):
    text[i] = str(text[i])
    text[i] = text[i].replace("\u200b","")
    text[i] = text[i].replace("\u200c","")
    text[i] = text[i].replace("\xa0"," ")
    text[i] = text[i].replace(":","")
    text[i] = text[i].replace("'","")
    text[i] = text[i].replace(".","")
    text[i] = text[i].replace(";","")
    text[i] = text[i].replace("|","")
    text[i] = text[i].replace(",","")
    text[i] = text[i].replace("Fact Check","")
punctuation_cleaning()
```
