##Importing libraries

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

##Scraping oneindia.com

```python
URL = "https://kannada.oneindia.com/fact-check"
r = requests.get(URL) 
soup = BeautifulSoup(r.content, 'html5lib')

soup.prettify()
```

##Extracting links from webpage 1

```python
lists = []
final_list = []

table = soup.find_all('div', attrs = {'class':'cityblock-title news-desc'})
for element in table:
  lists.append(element.find('a'))

#formatting
for link in lists:
  url = str(link)
  url = url.split('"')[1]
  temp_list=[]
  temp_list.append(url)
  final_list.append(temp_list)

print(final_list)
```

##Extracting links from webpages 2-17

---



```python
list_for_remaining_pages = []
for i in range(2,18):
  URL = "https://kannada.oneindia.com/fact-check/?page-no="+ str(i)
  r = requests.get(URL) 
  soup = BeautifulSoup(r.content, 'html5lib')
  table = soup.find_all('div', attrs = {'class':'cityblock-title news-desc'})
  for element in table:
    list_for_remaining_pages.append(element.find('a'))

#formatting    
for link in list_for_remaining_pages:
  url = str(link)
  url = url.split('"')[1]
  temp_list=[]
  temp_list.append(url)
  final_list.append(temp_list)

print(final_list)
```

##Checkpoint: CSV file containing links generated

```python
with open('oneindiaLinks.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(final_list)
```

##Reading csv file with links of all fake news articles

```python
df = pd.read_csv('oneindiaLinks.csv')
df.columns = ['url']
df['url'] = 'https://kannada.oneindia.com' + df['url']
```

##Scraping fake news headlines

```python
fake_news = []
for i in range(len(df)):
  URL = df.loc[i, 'url']
  r = requests.get(URL) 
  soup = BeautifulSoup(r.content, 'html5lib')
  table = soup.find('h1')
  fake_news.append(table)
```

##Extracting headlines between h1 tags

```python
final_list=[]

for i in fake_news:
  for j in i:
    temp_list = []
    temp_headline = str(j)
    temp_headline = temp_headline.split(":")
    if len(temp_headline)==1:
      temp_list.append(temp_headline[0])
    else:
      temp_list.append(temp_headline[1])
    final_list.append(temp_list)
    
print(final_list)
print("Length: ", len(final_list))
```

##Final output: CSV file containing fake news headlines generated

```python
with open('oneindiaFakeNews.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(final_list)
```
